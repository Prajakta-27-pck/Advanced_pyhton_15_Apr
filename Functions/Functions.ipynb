```python
def add(x:int|float, y:int|float):
    """This is used to add two numbers only"""
    assert isinstance(x,(int,float)),"Please insert x as int or float"
    assert isinstance(y,(int,float)),"Please insert x as int or float"
    z = x=y
    return z
```
